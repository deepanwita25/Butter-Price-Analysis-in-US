```python
import pymysql
```

```python
connection = pymysql.connect(host="localhost",
                        user = "root",
                        password = "root",
                        database = "Nass",
                        port = 3306,
                        autocommit = True)
```

```python
insert_query_butter = """
INSERT INTO `nass`.`butter`
(`date`, 
`month_name`, 
`month_no`, 
`year`, 
`product`, 
`product_desc`, 
`country`, 
`state_name`, 
`measure`, 
`unit`, 
`value`, 
`sector`, 
`source`)
VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
ON DUPLICATE KEY UPDATE
    value = VALUES(value),
    `updated_at` = IF(value<>VALUES(value), NOW(), `updated_at`)
"""
```

```python
cursor = connection.cursor()
```

```python
for _, row in butter_production_df.iterrows():
   cursor.execute(insert_query_butter,(
   row['date'],
   row['month_name'],
   row['month_no'],
   row['year'],
   row['product'],
   row['product_desc'],
   row['country'],
   row['state_name'],
   row['measure'],
   row['unit'],
   float(row['value']),
   row['sector'],
   row['source']))

connection.commit()
connection.close()
```
