```python
import pandas as pd
from pathlib import Path
```

```python
# Load the csv file
csv_file = 'data/US_milk.xlsx'
file_path = Path.cwd().parent/csv_file
file_path
```

```python
df = pd.read_excel(file_path)
df
```

```python
df['month'] = df['month'].str.lower()
```

```python
df['date'] = pd.to_datetime(df['year'].astype(str)+"-"+ df['month'], format ='%Y-%B')
```

```python
month_mapping = {
    'january':1,
    'february':2,
    'march':3,
    'april':4,
    'may':5,
    'june':6,
    'july':7,
    'august':8,
    'september':9,
    'october':10,
    'november':11,
    'december':12
    
}
```

```python
df['month_no'] = df['month'].map(month_mapping)
```

```python
def convert_cwt_to_kg(x):
    if x is None or x == "" or pd.isna(x):
        return None
    coeffcient = 100/45.359237
    new_value = float(x)*coeffcient
    return new_value
```

```python
# Filter out df containing only the wholesale price with USD/CWT
df_cwt = df[df['unit'] =='$/CWT'].copy()
df_cwt
```

```python
df_cwt['value']=df_cwt['value'].map(convert_cwt_to_kg)
```

```python
df_cwt['unit'] = '$/100kg'
df_cwt
```

```python
milk_df = pd.concat([df, df_cwt], ignore_index=True)
milk_df
```

```python
milk_df.columns.to_list()
```

```python
import pymysql
```

```python
connection = pymysql.connect(host = 'localhost',
                            user = 'root',
                            password = 'root',
                            database = 'clal',
                            port = 3306)
```

```python
cursor = connection.cursor()
```

```python
insert_sql_command = '''
    INSERT INTO `clal`.`milk`
    (`year`, `month`, `product_desc`, `unit`, `product`, `country`, `state_name`, `measure`, `source`, `value`, `date`, `month_no`)
    VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    ON DUPLICATE KEY UPDATE
        updated_at = IF(value<>VALUES(value), NOW(), updated_at),
        value = VALUES(value)
'''
```

```python
for _, row in milk_df.iterrows():
    cursor.execute(insert_sql_command,
                  (row['year'],
                   row['month'],
                   row['product_desc'],
                   row['unit'],
                   row['product'],
                   row['country'],
                   row['state_name'],
                   row['measure'],
                   row['source'],
                   float(row['value']),
                   row['date'],
                   row['month_no']
                  ))
connection.commit()
connection.close()
```
