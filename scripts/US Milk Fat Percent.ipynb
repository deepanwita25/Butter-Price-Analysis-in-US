```python
import pandas as pd
import pymysql
from pathlib import Path
```

```python
BASE_PATH = Path.cwd().parent
BASE_PATH
```

```python
# Load csv file
df = pd.read_csv(BASE_PATH/"data/MilkFat.csv")
df.drop(columns = ['Program', 'Geo Level', 'Domain', 'Domain Category'], inplace = True)
df
```

```python
df.columns = df.columns.str.lower()
```

```python
df_monthly = df[df['period']!='YEAR']
df_monthly
```

```python
month_mapping = {
    'JAN':1,
    'FEB':2,
    'MAR':3,
    'APR':4,
    'MAY':5,
    'JUN':6,
    'JUL':7,
    'AUG':8,
    'SEP':9,
    'OCT':10,
    'NOV':11,
    'DEC':12,
    }
```

```python
df_monthly['month_no'] = df_monthly['period'].map(month_mapping)
df_monthly['month_name'] = df_monthly['period'].str.lower()
```

```python
df_monthly.drop(['period', 'commodity'], axis = 1, inplace = True)
```

```python
df_monthly['country'] = 'US'
df_monthly['product'] = 'Milk'
df_monthly.rename(columns = {'data item': 'product_desc', 'state': 'state_name'}, inplace = True)
df_monthly['measure'] = 'Fat Content'
df_monthly['source'] = 'Nass'
df_monthly['sector'] = 'ANIMALS & PRODUCTS'
```

```python
df_monthly
```

```python
df_monthly['date'] = pd.to_datetime(df_monthly['year'].astype(str)+'-'+ df_monthly['month_name'], format = "%Y-%b")
df_monthly
```

```python
df_monthly.columns.to_list()
```

```python
df_monthly = df_monthly[['date', 'month_name', 'month_no', 'year', 
            'product', 'product_desc', 'country', 'state_name', 
            'measure', 'unit', 'value', 'sector', 'source']]
```

```python
df_monthly.isna().sum()
```

```python
df_monthly.dtypes
```

```python
# upload data to nass.butter table in the database
connection = pymysql.connect(host = 'localhost',
                            user = 'root',
                            password = 'root',
                            database = 'nass',
                            port = 3306)
```

```python
cursor = connection.cursor()
```

```python
insert_sql_command = '''
    INSERT INTO `nass`.`butter`
    (`date`, `month_name`, `month_no`, `year`, 
            `product`, `product_desc`, `country`, `state_name`, 
            `measure`, `unit`, `value`, `sector`, `source`)
    VALUES(%s, %s, %s, %s, %s,
            %s, %s, %s, %s,
            %s, %s, %s, %s)
    ON DUPLICATE KEY UPDATE
        updated_at = IF(value <> VALUES(value), NOW(), updated_at),
        value = VALUES(value)
'''
```

```python
for _ , row in df_monthly.iterrows():
    cursor.execute(insert_sql_command, (
    row['date'],
    row['month_name'],
    row['month_no'],
    row['year'],
    row['product'],
    row['product_desc'],
    row['country'],
    row['state_name'],
    row['measure'],
    row['unit'],
    row['value'],
    row['sector'],
    row['source']))

connection.commit()
connection.close()
```

```python
if connection.close:
    print("Database connection is closed")
elif connection.open:
    print("Database connection is still open")
```
