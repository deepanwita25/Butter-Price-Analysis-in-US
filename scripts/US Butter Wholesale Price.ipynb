```python
import pandas as pd
from pathlib import Path
import os
```

```python
root = Path(Path.cwd().parent)
root
```

```python
csv_file = "data/clal_butter_wholesale_price.csv"
```

```python
# Creating the file path for the csv file
file_path = root /csv_file
file_path
```

```python
# Loading the csv file with prices on Butter
df = pd.read_csv(file_path)
df.head(5)
```

```python
df.dtypes
```

```python
# Creating the month number column
month_number_mapping = {
    "January":1,
    "February":2,
    "March":3,
    "April":4,
    "May":5,
    "June":6,
    "July":7,
    "August":8,
    "September":9,
    "October":10,
    "November":11,
    "December":12
    }
```

```python
# Adding month_no column from month column
df['month_no'] = df['Month'].map(month_number_mapping)
df.tail(12)
```

```python
# Adding product column
df['product'] = 'butter'
df['country'] = 'US'
df['state_name'] = 'US TOTAL'
df['measure'] = 'Wholesale Price'
df['source'] = 'clal'
```

```python
df
```

```python
# Adding the date column
df['date'] = pd.to_datetime(df['year'].astype(str)+"-"+df['month_no'].astype(str)+"-"+"01")
df['value'] = df['Wholesale Price']
df = df.drop(columns=['Wholesale Price'], axis = 1)
df
```

```python
df.dtypes
```

```python
df['value'] = pd.to_numeric(df['value'].astype(str).str.replace(",", "", regex=False))
df
```

```python
df.dtypes
```

```python
df.columns.to_list()
```

The unit column has extra space which is not wanted..SO will strip off if any spaces exist in the column header

```python
df.columns= df.columns.str.strip()
```

```python
# Currently the set of values has unit USD/tonne
# Also need USD/kg
df_kg = df.copy()
df_kg['value'] = df_kg['value']/1000
df_kg['Unit'] = 'USD/kg'
```

```python
df_kg
```

```python
# Concating both the dfs
# Number of rows 71*2= 142
butter_price = pd.concat([df,df_kg], ignore_index = True)
butter_price.rename(columns = {'Month':'month',
                    'Product_desc':'product_desc',
                    'Unit':'unit'}, inplace= True)
```

```python
butter_price
```

```python
butter_price.isna().sum()
```

```python
butter_price.columns.to_list()
```

```python
# store data in the database
import pymysql
connection = pymysql.connect(host = 'localhost',
                            user = 'root',
                            password = 'root',
                            database = 'nass',
                            port = 3306,
                            autocommit = True)
```

```python
cursor = connection.cursor()
```

```python
insert_sql_command = '''
    INSERT INTO `clal`.`wholesale_price`
    (`year`, `month`, `product_desc`,`unit`,`month_no`,`product`,`country`, `state_name`,`measure`,`source`,`date`,`value`)
    VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
        ON DUPLICATE KEY UPDATE
        `updated_at` = IF(value <> VALUES(value), NOW(), updated_at),
        value = VALUES(value)
'''
```

```python
for _, row in butter_price.iterrows():
    cursor.execute(insert_sql_command,
                   (
                        row['year'],
                        row['month'],
                        row['product_desc'],
                        row['unit'],
                        row['month_no'],
                        row['product'],
                        row['country'],
                        row['state_name'],
                        row['measure'],
                        row['source'],
                        row['date'],
                        float(row['value'])
                   )
                  )
connection.commit()
connection.close()
```
